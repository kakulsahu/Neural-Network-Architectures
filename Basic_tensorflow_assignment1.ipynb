```python
tensor=tf.random.uniform(shape=(3,3),minval=0,maxval=1)
```

```python
print("Random Tensor:")
print(tensor)
```

# Perform basic tensor operations including matrix addition, multiplication, and transpose.

```python
A = tf.random.uniform((3, 3), minval=0, maxval=1)
B = tf.random.uniform((3, 3), minval=0, maxval=1)
```

```python
print("Tensor A:",A)
print("Tensor B:",B)
```

```python
# Matrix Addition
add_result = tf.add(A, B)
print("Matrix Addition (A + B):")
print(add_result)
```

```python
# Multiplication
mul_result = tf.matmul(A, B)
print("Matrix Multiplication (A x B):")
print(mul_result)
```

```python
# Transpose
transpose_A = tf.transpose(A)
print("Transpose of Tensor A:")
print(transpose_A)
```

# Convert the resulting tensor into a NumPy array and then reshape it into a (1, 9) tensor.

```python
tensor=tf.random.uniform(shape=(3,3),minval=0,maxval=1)
print("Original Tensor:",tensor)
```

```python
numpy_array=tensor.numpy()
print("Converted Numpy array:",numpy_array)
```

```python
reshaped_tensor=tf.reshape(tensor,(1,9))
print("Reshaped Tensor:",reshaped_tensor)
```
